### Exercise 5: Subplot Layout

```python
# TODO: Create a 2x2 subplot layout.
# Plot a line chart in the first subplot, a bar chart in the second, a scatter plot in the third, and a pie chart in the fourth.
import matplotlib.pyplot as plt

fig, axs = plt.subplots(2, 2)

axs[0, 0].plot([1, 2, 3, 4], [10, 20, 25, 30])
axs[0, 1].bar(['A', 'B', 'C', 'D'], [10, 20, 25, 30])
axs[1, 0].scatter([1, 2, 3, 4], [10, 20, 25, 30])
axs[1, 1].pie([40, 25, 20, 15], labels=['Python', 'Java', 'C++', 'JavaScript'], autopct='%1.1f%%')

plt.tight_layout()
plt.show()
```
